```python
def first_line(fn):
    #['time_updated', 'team_name', 'game_date', 'players', 'sport_type']
    l = []
    with open(fn, newline='', encoding='utf-8') as csvfile:
        reader = csv.reader(csvfile, delimiter=',', quotechar='"') 
        count = 0
        for row in reader:
            if count:
                date = datetime.strptime(str(row[2][2:-6]), '%y-%m-%dT%H:%M:%S')
                d = json.loads(row[3])
                sport = row[4]
                if sport =='SOCCER':
                    l.append( (date, d) )
                    for key in d.keys():
                        print(key)
            count+=1
    return l
            
l = first_line('prizepicks_dec.csv')    
```

```python
def partition(l):
    id_count = 0
    id_to_object = {}
    players = {}
    stat_types = {}
    projection_types = {}
    positions = {}
    
    for date, d in l:
        for player in d:
            projection_data = d[player]['L'][0]['M']
            
            id_to_object[id_count] = projection_data
            
            stat_type = projection_data['stat_type']['S']
            projection_type = projection_data['projection_type']['S']
            position = projection_data['position']['S']
            
            players[player] = players.get(player, []) + [id_count]
            stat_types[stat_type] = stat_types.get(stat_type, []) + [id_count]
            projection_types[projection_type] = projection_types.get(projection_type, []) + [id_count]
            positions[position] = positions.get(position, []) + [id_count]
            
            id_count += 1
            
    return id_to_object, players, stat_types, projection_types, positions
        
id_to_object, players, stat_types, projection_types, positions = partition(l)
```

```python
def filter_projections(player_name = None, stat_type = None, projection_type = None, position = None):
    #id_to_object, players, stat_types, projection_types, positions
    ids = set(id_to_object.keys())
    
    if player_name:
        ids = ids.intersection(set(players.get(player_name, [])))
    if stat_type:
        ids = ids.intersection(set(stat_types.get(stat_type, [])))
    if projection_type:
        ids = ids.intersection(set(projection_types.get(projection_type, [])))
    if position:
        ids = ids.intersection(set(positions.get(position, [])))
        
    projections = []
    for i in ids:
        projections.append(id_to_object[i])
        
    return projections

    
filter_projections(player_name = 'Harry Kane')
```
