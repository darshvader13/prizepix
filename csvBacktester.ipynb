```python
offset = 0
shortname_teams = {}
teams = {}
```

```python
while True:
    test_url = api_url + '?limit=500&offset=' + str(offset)
    test_response = requests.get(test_url, headers=headers)
    test_response_json = test_response.json()
    for team in test_response_json.get('teams', []):
        shortname_teams[ team['shortName'] ] = team['id']
        teams[ team['name'] ] = team['id']
    count = test_response_json.get('count', 0)
    if count != 500:
        break
    offset += 500
```
